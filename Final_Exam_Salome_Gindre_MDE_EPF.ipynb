5. create a function that takes a strain and returns the (conditional) probability of getting an A under the assumption that the precedent letter is C. Do the same for getting A under the assumption that before you have a G. 


We take the stain of the beginning : seq_record

```python

def probability_A_after_C(ListOfNucleo):
    countA=0
    countTotal=0
    for i in range(0,len(ListOfNucleo)-1):
        # we check first is the letter is C
        if(ListOfNucleo[i] == 'C'): 
            countTotal+=1
            # then if the next letter is an A
            if (ListOfNucleo[i+1]=='A'):
                countA+=1
    return countA/countTotal 

def probability_A_after_G(ListOfNucleo):
    countA=0
    countTotal=0
    for i in range(0,len(ListOfNucleo)-1):
        if(ListOfNucleo[i] == 'G'):
            countTotal+=1
            if(ListOfNucleo[i+1]=='A'):
                countA+=1
    return countA/countTotal 
```

We then compute the conditional probabilities and we obtain : 

```python
print("The conditional probability that we find an A after a C in the strain is ",probability_A_after_C(seq_record))
print("The conditional probability that we find an A after a C in the strain is ",probability_A_after_G(seq_record))
```
