```python
#make the requests to get the csv file
import requests

total_deaths_url = "https://covid.ourworldindata.org/data/ecdc/total_deaths.csv"
new_deaths_url = "https://covid.ourworldindata.org/data/ecdc/new_deaths.csv"
csv_file = requests.get(total_deaths_url)
csv_file2 = requests.get(new_deaths_url)
open("total_deaths.csv", "wb").write(csv_file.content)
open("new_deaths.csv", "wb").write(csv_file2.content)

import pandas as pd

csvfname = "total_deaths.csv"
csvfname2 = "new_deaths.csv"
df = pd.read_csv(csvfname, index_col="date", header=0, skipinitialspace=True)
df2 = pd.read_csv(csvfname2, index_col="date", header=0, skipinitialspace=True)

df=df['2020-01-01':'2020-04-22']
df2=df2['2020-03-10':'2020-04-22']
#df.head(100)
#df[:10]
df2[:10]
                        
```

```python
print(f"{df.shape} \n" )
print(f"{df.dtypes} \n")
print(f"{df.isna().sum()}")
```

```python
#plot some curves of different countries
import matplotlib.pyplot as plt
plt.figure(0)
df['United States'].plot(figsize=(12,7), title ='Total number of deaths for different countries')
df['China'].plot()
df['France'].plot()
df['Italy'].plot()
df['Germany'].plot()
df['Sweden'].plot()
plt.legend()
plt.show()
```
