## Eigenvalues, Eigenvectors and Eigenspaces

```
import numpy as np
from numpy import linalg as LA 
import matplotlib.pyplot as plt

# 1. 
# compute the eigenvalues, eignevectors of the following matrix A 
# Real matrix possessing complex e-values and e-vectors; note that the e-values are complex conjugates of each other.
A = np.array([[-2,1],[3,5]])
w,e = LA.eig(A)

w;e
print (e,w)
#w=eigenvalues 

# 2. 
# Draw the vector field associated to A, the eigenspaces of A 
# the function returns coordinate matrices from coordinate vectors.

#x = np.linspace(0, 1)
#y = np.linspace(0, 1)
#xv, yv = np.meshgrid(x, y) 

#print(xv,yv)

x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
xx, yy = np.meshgrid(-2*x+y, 3*x+5*y)

fig,ax = plt.subplots()
q = ax.quiver(x,y,xx,yy)
ax.quiverkey(q, X=0.3, Y=1.1, U=10, label='Vector field', labelpos='E')
plt.show()

#quiver : champ de vecteur 

#h = plt.contourf(x,y)
plt.show()

# 3.



```
