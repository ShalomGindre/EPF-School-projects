# Moving average processess (MA)

We consider that past noises are influencing current values maybe with a different weight. 

MA process are maybe not too frequent in the nature but in the tradinf field it can make sense. 

>different test 

> What we want 
xt = Zt + 0.72t-1 + 0.2Zt-2

**Example**

```
dim = 1000
MA_2 = np.zeros(dim)
```

```
np.random.seed(1)

noise = np.random.normal(0,1,1000)
```

```
for i in range(2, dim):
  MA_2[i] = 0.7*noise[i-1] + 0.2*noise[i-1] + noise[i]
```

```
plt.plot(MA_2)
```

We can see it's not only noise 

```
fig = plot_acf(MA_2)
```

# Autoregressive Processes 

AR are not necesseray stationnary when it is defined as random walf.
it could be random shock or innovation 

```
dim = 10000
noise = np.random.normal(0,1,dim )
X = np.zeros(dim)
```

```
for i in range(1, dim):
  X[i] = noise[i] + 0.8*X[i-1]
```

```
fig = plt.plot(X[:500])
```

```
fig = plot_acf(X)
```

Weknow taht the model is correct because of the trend on the ACF. There is a link with the day before (and again, and again)

We loose the correlation at a certain point. It lookks stationnary on the plot to the 500 points. 

AR2

```
dim = 10000
noise = np.random.normal(0,1,dim )
AR2 = np.zeros(dim)

for i in range(1, dim):
  AR2[i] = noise[i] + 0.2*X[i-1]
```

ACF is up down between pos and neg correlation (between t and t+1)

If you increase at one pint, the next one will decrease 

If is the look of an AR process. You can find q from the ACF
